# Stop Words
Words like "a" and "the" appear so frequently that they don't require tagging as thoroughly as nouns, verbs and modifiers. We call these *stop words*, and they can be filtered from the text to be processed. spaCy holds a built-in list of some 305 English stop words.

```
import spacy.cli
spacy.cli.download('en_core_web_sm')
```

```
import spacy
nlp = spacy.load('en_core_web_sm')
```

```
print(nlp.Defaults.stop_words)
```

```
len(nlp.Defaults.stop_words)
```

```
nlp.vocab['first'].is_stop
```

```
nlp.vocab['Nerchuko'].is_stop
```

#Adding a stop word to the vocabulary

```
nlp.Defaults.stop_words.add('nerchuko')

nlp.vocab['nerchuko'].is_stop = True
```

```
nlp.vocab['nerchuko'].is_stop
```

```
len(nlp.Defaults.stop_words)
```

#Removing a stop word from the vocabulary

```
nlp.Defaults.stop_words.remove('nerchuko')
nlp.vocab['nerchuko'].is_stop = False
```

```
len(nlp.Defaults.stop_words)
```

```
nlp.vocab['nerchuko'].is_stop
```

```
nlp.Defaults.stop_words.remove('only')
nlp.vocab['only'].is_stop = False
```

```
len(nlp.Defaults.stop_words)
```

```
nlp.vocab['only'].is_stop
```
